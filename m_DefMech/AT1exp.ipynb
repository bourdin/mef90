```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def a(s,b):
    if b == 0:
        return 1-s
    else:
        return 1 + (np.e**(-b*s) - 1) / (1 - np.e**(-b))
def ap0(b):
    if b == 0:
        return 1
    else:
        return -b / (1-np.e**(-b))
def sigmac(b):
    return np.sqrt(-2./ap0(b))
```

```python
N = 101
S = np.linspace(0,1,N)
for b in np.linspace(0,4,5):
    plt.plot([a(s,b) for s in S],label = "b = {}".format(b))
plt.legend(loc=0)
plt.show()
```

```python
B = np.linspace(0.01,10,11)
plt.plot(B,[sigmac(b) for b in B])
plt.grid()
plt.show()
```
